```python
from tkinter import *
import random as rn
import time as time

root = Tk()

root.config()
root.title('Awesome Rock, Paper, Scissors Game')
root.iconbitmap('rockicon.ico')
root.geometry('400x400')

moves = ['Rock', 'Paper', 'Scissors']
comp_wins = 0
player_wins = 0
player_pick = ' '
computer_pick = ' '
   
def rock():
    global player_pick
    player_pick = 'Rock'
    print(player_pick)
    win_check()
    
def paper():
    global player_pick
    player_pick = 'Paper'
    print(player_pick)
    win_check()
    
def scissors():
    global player_pick
    player_pick = 'Scissors'
    print(player_pick)
    win_check()
    
def win_check():
    global computer_pick
    computer_pick = rn.choice(moves)
    
    if player_pick == computer_pick:
        round_label.config(text='You picked {}, the computer also picked {}. Please pick again.'.format(player_pick, computer_pick))

  #player wins
    elif player_pick == 'Rock' and computer_pick == 'Scissors':
        player_win()
    elif player_pick == 'Paper' and computer_pick == 'Rock':
        player_win()
    elif player_pick == 'Scissors' and computer_pick == 'Paper':
        player_win()

  #computer wins
    elif computer_pick == 'Rock' and player_pick == 'Scissors':
        comp_win()
    elif computer_pick == 'Paper' and player_pick == 'Rock':
        comp_win()
    elif computer_pick == 'Scissors' and player_pick == 'Paper':
        comp_win()

def player_win():
    global player_wins
    round_label.config(text='You picked {}, the computer picked {}. You won this round!.'.format(player_pick, computer_pick)) 
    player_wins += 1
    score()
    
def comp_win():
    global comp_wins
    round_label.config(text='You picked {}, the computer picked {}. Computer wins this round!.'.format(player_pick, computer_pick))
    comp_wins += 1
    score()
    
def score():
    score_counter.config(text='Computer - {}, Player - {}'.format(comp_wins, player_wins))
    if comp_wins == 2:
        win_label.config(text='Computer wins!')
    elif player_wins == 2:
        win_label.config(text='Congratulations! You win!')

def restart():
    global comp_wins
    global player_wins
    global player_pick
    global computer_pick
    comp_wins = 0
    player_wins = 0
    player_pick = ' '
    computer_pick = ' '
    score()
    round_label.config(text=' ')
    win_label.config(text=' ')
    
    
rock_image = PhotoImage(file = 'rock.gif')
scissors_image = PhotoImage(file = 'scissors.gif')
paper_image = PhotoImage(file = 'paper.gif')


title_label = Label(root, text='Rock, Paper, Scissors vs Computer')
game_label = Label(root, text='Best 2 out of 3')
score_counter = Label(root, text='Computer - {}, Player - {}'.format(comp_wins, player_wins))
round_label = Label(root, text=' ')
win_label = Label(root, text=' ')

rock_button = Button(root, image=rock_image, command=rock)
scissors_button = Button(root, image=scissors_image, command=scissors)
paper_button = Button(root, image=paper_image, command=paper)
restart_button = Button(root, text='Restart', bg='Black', fg='White', command=restart)

restart_button.grid(row=10, column=3)
win_label.grid(row=6, column=1, columnspan=3)
round_label.grid(row=5, column=1, columnspan=3)
score_counter.grid(row=4, column=1, columnspan=3)
title_label.grid(row=1, column=1, columnspan=3)
game_label.grid(row=2, column=1, columnspan=3)
rock_button.grid(row=3, column=1)
paper_button.grid(row=3, column=2)
scissors_button.grid(row=3, column=3)

root.mainloop()

```
